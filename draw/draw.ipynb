```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 
df = pd.read_csv('results.csv')

# 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 
plt.figure(figsize=(12, 6))

# 
metrics = ['Accuracy', 'Precision', 'Recall', 'F1_Score']
types = df['Type'].values  # 'Type'
x = np.arange(len(types))
width = 0.2

# 
for i, metric in enumerate(metrics):
    plt.bar(x + width * (i - len(metrics)/2 + 0.5), 
           df[metric].values, 
           width, 
           label=metric.replace('_', ' '))

# 
plt.ylabel(' (%)')
plt.title('')
plt.xticks(x, types)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# 
plt.tight_layout()

# 
plt.savefig('performance_comparison.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 
df = pd.read_csv('results.csv')

# 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 
plt.figure(figsize=(10, 6))

# 
types = df['Type'].values
coverage_data = df['Unknown_covery'].values

# 
bars = plt.bar(types, coverage_data, 
               width=0.6,  # 
               color='#1f77b4',  # 
               edgecolor='black',  # 
               linewidth=1)  # 

# 
plt.ylabel('Unknown Coverage (%)')
plt.title('Different Types Unknown Coverage Comparison')
plt.grid(True, linestyle='--', alpha=0.3, axis='y')

# 
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.1f}%',
            ha='center', va='bottom',
            fontsize=10)

# y
plt.ylim(0, 110)
plt.yticks(np.arange(0, 101, 10))

# 
plt.tight_layout()

# 
plt.savefig('unknown_coverage_comparison.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 
df = pd.read_csv('results.csv')

# 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 
plt.figure(figsize=(10, 6))

# 
types = df['type'].values
far_data = df['false_alarm_rate'].values * 100

# 
bars = plt.bar(types, far_data, 
               width=0.6,
               color='#ff7f0e',  # 
               edgecolor='black',
               linewidth=1)

# 
plt.ylabel('False Alarm Rate (%)')
plt.title('Different Types False Alarm Rate Comparison')
plt.grid(True, linestyle='--', alpha=0.3, axis='y')

# 
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.1f}%',
            ha='center', va='bottom',
            fontsize=10)

# y
plt.ylim(0, max(far_data) * 1.2)  # 20%
plt.yticks(np.arange(0, max(far_data) * 1.2, 5))

# 
plt.tight_layout()

# 
plt.savefig('far_comparison.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 
df = pd.read_csv('results.csv')

# 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 
plt.figure(figsize=(10, 6))

# 
types = df['type'].values
detector_data = df['num_detectors'].values

# 
bars = plt.bar(types, detector_data, 
               width=0.6,
               color='#2ca02c',  # 
               edgecolor='black',
               linewidth=1)

# 
plt.ylabel('Number of Detectors')
plt.title('Different Types Detector Number Comparison')
plt.grid(True, linestyle='--', alpha=0.3, axis='y')

# 
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height)}',  # 
            ha='center', va='bottom',
            fontsize=10)

# y
plt.ylim(0, max(detector_data) * 1.1)  # 10%
plt.yticks(np.arange(0, max(detector_data) * 1.1, 2000))

# 
plt.tight_layout()

# 
plt.savefig('detector_number_comparison.png', bbox_inches='tight', dpi=300)
plt.show()
```
