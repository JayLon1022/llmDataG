```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取数据
df = pd.read_csv('results.csv')

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 设置图形大小
plt.figure(figsize=(12, 6))

# 准备数据
metrics = ['accuracy', 'precision', 'recall', 'f1_score']
types = df['type'].values
x = np.arange(len(types))
width = 0.2

# 绘制柱状图
for i, metric in enumerate(metrics):
    plt.bar(x + width * (i - len(metrics)/2 + 0.5), 
           df[metric].values * 100,  # 转换为百分比
           width, 
           label=metric.replace('_', ' ').title())

# 设置图表格式
plt.ylabel('百分比 (%)')
plt.title('不同类型的性能指标比较')
plt.xticks(x, types)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# 调整布局
plt.tight_layout()

# 保存图表
plt.savefig('performance_comparison.svg', bbox_inches='tight', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取数据
df = pd.read_csv('results.csv')

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 设置图形大小
plt.figure(figsize=(10, 6))

# 准备数据
types = df['type'].values
coverage_data = df['unknown_coverage'].values * 100

# 绘制柱状图
bars = plt.bar(types, coverage_data, 
               width=0.6,  # 调整柱子宽度
               color='#1f77b4',  # 设置颜色
               edgecolor='black',  # 添加边框
               linewidth=1)  # 边框宽度

# 设置图表格式
plt.ylabel('Unknown Coverage (%)')
plt.title('Different Types Unknown Coverage Comparison')
plt.grid(True, linestyle='--', alpha=0.3, axis='y')

# 为每个柱子添加数值标签
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.1f}%',
            ha='center', va='bottom',
            fontsize=10)

# 调整y轴范围，留出一些空间显示数值标签
plt.ylim(0, 110)
plt.yticks(np.arange(0, 101, 10))

# 调整布局
plt.tight_layout()

# 保存图表
plt.savefig('unknown_coverage_comparison.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取数据
df = pd.read_csv('results.csv')

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 设置图形大小
plt.figure(figsize=(10, 6))

# 准备数据
types = df['type'].values
far_data = df['false_alarm_rate'].values * 100

# 绘制柱状图
bars = plt.bar(types, far_data, 
               width=0.6,
               color='#ff7f0e',  # 使用不同的颜色
               edgecolor='black',
               linewidth=1)

# 设置图表格式
plt.ylabel('False Alarm Rate (%)')
plt.title('Different Types False Alarm Rate Comparison')
plt.grid(True, linestyle='--', alpha=0.3, axis='y')

# 为每个柱子添加数值标签
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.1f}%',
            ha='center', va='bottom',
            fontsize=10)

# 调整y轴范围
plt.ylim(0, max(far_data) * 1.2)  # 留出20%的空间显示标签
plt.yticks(np.arange(0, max(far_data) * 1.2, 5))

# 调整布局
plt.tight_layout()

# 保存图表
plt.savefig('far_comparison.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取数据
df = pd.read_csv('results.csv')

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 设置图形大小
plt.figure(figsize=(10, 6))

# 准备数据
types = df['type'].values
detector_data = df['num_detectors'].values

# 绘制柱状图
bars = plt.bar(types, detector_data, 
               width=0.6,
               color='#2ca02c',  # 使用另一种颜色
               edgecolor='black',
               linewidth=1)

# 设置图表格式
plt.ylabel('Number of Detectors')
plt.title('Different Types Detector Number Comparison')
plt.grid(True, linestyle='--', alpha=0.3, axis='y')

# 为每个柱子添加数值标签
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height)}',  # 整数格式
            ha='center', va='bottom',
            fontsize=10)

# 调整y轴范围
plt.ylim(0, max(detector_data) * 1.1)  # 留出10%的空间显示标签
plt.yticks(np.arange(0, max(detector_data) * 1.1, 2000))

# 调整布局
plt.tight_layout()

# 保存图表
plt.savefig('detector_number_comparison.png', bbox_inches='tight', dpi=300)
plt.show()
```
