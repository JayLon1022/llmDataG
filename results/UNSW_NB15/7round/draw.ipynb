```python
# LLM内各指标对比

import matplotlib.pyplot as plt
import numpy as np

# Data preparation
types = ['A', 'B', 'D', 'E', 'F', 'G', 'R', 'S', 'W']
coverage = [100.00, 99.83, 98.19, 89.68, 74.94, 99.69, 92.33, 100.00, 100.00]
accuracy = [90.48, 89.87, 90.24, 90.00, 89.87, 91.27, 91.25, 92.01, 90.66]
precision = [90.51, 90.93, 91.18, 90.73, 91.51, 91.44, 91.41, 91.44, 91.29]
recall = [90.44, 88.58, 89.10, 89.10, 87.90, 91.06, 91.06, 92.70, 89.90]
f1_score = [90.48, 89.74, 90.13, 89.91, 89.67, 91.25, 91.23, 92.06, 90.59]
fpr = [9.48, 8.84, 8.62, 9.10, 8.16, 8.52, 8.56, 8.68, 8.58]

# Set positions for bar chart
x = np.arange(len(types))
width = 0.15

# Create figure
plt.figure(figsize=(15, 7))

# Draw bar charts
plt.bar(x - width*2, coverage, width, label='Unknown Coverage', color='purple')
plt.bar(x - width, accuracy, width, label='Accuracy', color='skyblue')
plt.bar(x, precision, width, label='Precision', color='lightgreen')
plt.bar(x + width*0.5, recall, width, label='Recall', color='salmon')
plt.bar(x + width*1.5, f1_score, width, label='F1 Score', color='lightblue')
plt.bar(x + width*2, fpr, width, label='False Positive Rate', color='orange')

# Set figure properties
plt.xlabel('Unknown Type')
plt.ylabel('Percentage (%)')
plt.title('LLM_Based UNSW-NB15 Performance Metrics ')
plt.xticks(x, types)
plt.legend()

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()
# Save figure
plt.savefig('LLM_Based_UNSW-NB15_Performance_Metrics.png', dpi=300, bbox_inches='tight')
```

```python
# LLM内检测器数量对比
import matplotlib.pyplot as plt

# Data preparation
types = ['A', 'B', 'D', 'E', 'F', 'G', 'R', 'S', 'W']
detectors = [19792, 19791, 19863, 19332, 19523, 19651, 14237, 20656, 19969]

# Create figure
plt.figure(figsize=(12, 6))

# Draw bar chart
bars = plt.bar(types, detectors, color='skyblue')

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom')

# Set figure properties
plt.xlabel('Type')
plt.ylabel('Number of Detectors')
plt.title('Comparison of Detector Counts by Type')

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

# Save figure
plt.savefig('detector_count_comparison.png', dpi=300, bbox_inches='tight')
```
