```python
# 
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
types = ['A', 'B', 'D', 'E', 'F', 'G', 'R', 'S', 'W']
accuracy = [50.08, 53.99, 51.66, 51.48, 50.30, 50.99, 49.99, 50.19, 50.79]
precision = [100.00, 99.50, 99.40, 96.25, 100.00, 94.59, 33.33, 100.00, 100.00]
recall = [0.16, 8.02, 3.34, 3.08, 0.60, 2.10, 0.02, 0.38, 1.58]
f1_score = [0.32, 14.84, 6.46, 5.97, 1.19, 4.11, 0.04, 0.76, 3.11]
fpr = [0.00, 0.04, 0.02, 0.12, 0.00, 0.12, 0.04, 0.00, 0.00]
coverage = [29.92, 9.45, 1.11, 0.58, 0.86, 18.70, 0.14, 19.79, 37.36]

# Set bar positions
x = np.arange(len(types))
width = 0.13

# Create figure
plt.figure(figsize=(15, 7))

# Plot bars
plt.bar(x - width*2.5, accuracy, width, label='Accuracy', color='skyblue')
plt.bar(x - width*1.5, precision, width, label='Precision', color='lightgreen')
plt.bar(x - width*0.5, recall, width, label='Recall', color='salmon')
plt.bar(x + width*0.5, f1_score, width, label='F1 Score', color='lightblue')
plt.bar(x + width*1.5, fpr, width, label='FPR', color='orange')
plt.bar(x + width*2.5, coverage, width, label='Unknown Coverage', color='purple')

# Set figure properties
plt.xlabel('Detector Type')
plt.ylabel('Percentage (%)')
plt.title('Performance Metrics Comparison of Different Detector Types (VD Method)')
plt.xticks(x, types)
plt.legend()

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

# Save figure
plt.savefig('baseline_vd_metrics_comparison.png', dpi=300, bbox_inches='tight')
```

```python
# 
import matplotlib.pyplot as plt

# Data preparation
types = ['A', 'B', 'D', 'E', 'F', 'G', 'R', 'S', 'W']
detectors = [30, 20, 30, 50, 20, 50, 10, 30, 20]

# Create figure
plt.figure(figsize=(12, 6))

# Draw bar chart
bars = plt.bar(types, detectors, color='skyblue')

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom')

# Set figure properties
plt.xlabel('Type')
plt.ylabel('Number of Detectors')
plt.title('Comparison of Detector Counts by Type')

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

# Save figure
plt.savefig('detector_count_comparison.png', dpi=300, bbox_inches='tight')
```
